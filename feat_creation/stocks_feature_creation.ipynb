```python
#Import modules
import os
import glob
import pandas as pd
#!pip install finta
from finta import TA
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid', context='talk', palette='Dark2')
```

```python
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
ticker_csv_files = glob.glob(os.path.join(path, "ticker_*.csv"))
print(ticker_csv_files)
```

```python
stock_list = []
for f in ticker_csv_files:
    stock_list.append(f.split("r_")[-1].split('.')[0])

print(stock_list)

stock_df = dict(map(lambda i,j : (i,j), stock_list,ticker_csv_files))
```

```python
stocks = {}
# loop over the list of csv files
for i, f in stock_df.items():
    # read the csv file
    stocks[i] = pd.read_csv(f, parse_dates = ['Unnamed: 0'])
    stocks[i].rename(columns={'Unnamed: 0':'date'}, inplace=True)
    stocks[i].set_index('date', inplace=True)
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("ticker_")[-1])
      
    # print the content
    print('Content:')
    display(stocks[i])
    print()
    
```

```python
close_stocks = {}
for i, data in stocks.items():
    close_stocks[i] = {}
    close_stocks[i]['close'] = data['close']
    temp = close_stocks[i]['close'][:40]
    
    temp.plot(title=i)
    plt.show()
#print(close_stocks)
```

```python
smooth_closes = {}
for i, data in close_stocks.items():
    smooth_closes[i] = {}
    smooth_closes[i]['close'] = data['close'].ewm(alpha=0.7).mean()
    temp = smooth_closes[i]['close'][:40]
    
    temp.plot(title=i)
    plt.show()
#print(smooth_closes)
```

```python
for i, data in stocks.items():
    stocks[i]['close'] = smooth_closes[i]['close']
```

```python
def get_indicator_data(data, indicators):
    """
    Function that uses the finta API to calculate technical indicators used as the features
    :return:
    """

    for indicator in indicators:
        ind_data = eval('TA.' + indicator + '(data)')
        if not isinstance(ind_data, pd.DataFrame):
            ind_data = ind_data.to_frame()
        data = data.merge(ind_data, left_index=True, right_index=True)
    data.rename(columns={"14 period EMV.": '14 period EMV'}, inplace=True)

    # Also calculate moving averages for features
    data['ema60'] = data['close'] / data['close'].ewm(60).mean()
    data['ema20'] = data['close'] / data['close'].ewm(20).mean()
    data['ema10'] = data['close'] / data['close'].ewm(10).mean()
    data['ema5'] = data['close'] / data['close'].ewm(5).mean()

    # Instead of using the actual volume value (which changes over time), we normalize it with a moving volume average
    data['ema5Vol'] = data['volume'] / data['volume'].ewm(5).mean()
    
    return data
```

```python
# List of symbols for technical indicators
indicators = ['BBANDS', 'PIVOT_FIB', 'RSI', 'MACD', 'STOCH','STOCHRSI', 'ADL', 'ATR', 'MOM', 'MFI', 'ROC', 'OBV', 'CCI', 'EMV', 'VORTEX']
technicals = {}
for i, data in stocks.items():
    technicals[i] = get_indicator_data(data, indicators)
```
